# EDA & Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Salaries.csv", index_col=0)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Year'] = pd.to_datetime(df['Year']).dt.year
df
```

```python
df.info()
```

```python
df.drop_duplicates()
```

```python
df.isna().sum()
```

Then we will drop Notes and Status columns as they don't have any Value

```python
df.drop(['Notes','Status'],axis=1,inplace=True)
```

We Will delete unvalid Values for the numerical columns

```python
invalid_rows = df[(df['BasePay'] < 0) |( df['OtherPay'] < 0) |( df['OvertimePay'] < 0) | (df['Benefits'] < 0) | (df['TotalPay'] < 0) | (df['TotalPayBenefits'] < 0)]
df.drop(invalid_rows.index, axis=0, inplace=True)
df.describe()
```

```python
df.isna().sum()
```

We will drop the rows contain NA in Overtime and other pay columns as they are only 4 rows

```python
columns_to_check = ['OvertimePay', 'OtherPay']
df.dropna(subset=columns_to_check, inplace=True)
```

We will fill the rest NAN values with mean values of this columns because the nan values has big proportion of data that we can't lose so the mean will be the appropriate value to use to fill NANs

```python
columns_to_fill = ['Benefits', 'BasePay']

for column in columns_to_fill:
    df[column].fillna(df[column].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
df.Agency.value_counts()
```

```python
df.drop('Agency',axis=1,inplace=True)
```

```python
df.describe()
```

# Feature Engineering

```python
df['TotalPayBenefits']=df['TotalPay']+df['Benefits']
```

```python
df[['Job Title', 'Department']] = df['JobTitle'].str.extract(r'([^()]+)\s?\(([^()]+)\)')

df['Department'].fillna('Management', inplace=True)

df['Job Title'].fillna(df['JobTitle'], inplace=True)

df.drop('JobTitle', axis=1, inplace=True)

df
```

```python
df.Department.value_counts()
```

```python
department_mapping = {
    'CIVIL & CRIMINAL': 'Civil & Criminal',
    'Cvl&Crmnl': 'Civil & Criminal',
    'CVL&CRMNL': 'Civil & Criminal',
    'Civil & Criminal': 'Civil & Criminal',
    'CIVIL/CRIMINAL': 'Civil & Criminal',
    'Civil/Criminal': 'Civil & Criminal',
    'Fire Department': 'Fire',
    'Fire Dept': 'Fire',
    'FIRE DEPT': 'Fire',
    'FIRE DEPARTMENT': 'Fire',
    'POLICE DEPARTMENT': 'Police',
    'Police Department': 'Police',
    'SEASONAL': 'Seasonal',
    'Seasonal': 'Seasonal'
}

df['Department'] = df['Department'].str.upper().replace(department_mapping)

plt.figure(figsize=(10, 8))

department_counts = df['Department'].value_counts()
labels = department_counts.index

colors = sns.color_palette('pastel')[0:len(labels)]

plt.pie(department_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Proportion of Employees in Different Departments (excluding Management)')
plt.axis('equal')

plt.legend(labels, loc='upper right', bbox_to_anchor=(1, 0.9))

plt.show()
```

```python
filtered_df = df[(df['Department'] != 'MANAGEMENT') & (df['Department'].notna())]

plt.figure(figsize=(10, 8))

department_counts = filtered_df['Department'].value_counts()
labels = department_counts.index

colors = sns.color_palette('pastel')[0:len(labels)]

plt.pie(department_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Proportion of Employees in Different Departments (excluding Management)')
plt.axis('equal')

plt.legend(labels, loc='upper right', bbox_to_anchor=(1, 0.9))

plt.show()
```

# Aggregations

```python
mean_pay_by_department=pd.DataFrame(df.groupby("Department")['TotalPayBenefits'].mean())
mean_pay_by_department
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=mean_pay_by_department, x='Department', y='TotalPayBenefits', palette='viridis')
plt.title('Mean Total Pay + Benefits by Department')
plt.xlabel('Department')
plt.ylabel('Mean Total Pay + Benefits')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
mean_benefits_by_department=pd.DataFrame(df.groupby("Department")['Benefits'].mean())
mean_benefits_by_department
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=mean_benefits_by_department, x='Department', y='Benefits', palette='magma')
plt.title('Mean Benefits by Department')
plt.xlabel('Department')
plt.ylabel('Mean Benefits')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
mean_basepay_by_department=pd.DataFrame(df.groupby("Department")['BasePay'].mean())
mean_basepay_by_department
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=mean_basepay_by_department, x='Department', y='BasePay', palette='plasma')
plt.title('Mean BasePay by Department')
plt.xlabel('Department')
plt.ylabel('Mean BasePay')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```
